```python
import pandas as pd
import requests
import sqlite3
import matplotlib.pyplot as plt

# Pobieramy dane z API - dodajemy fields, żeby uniknąć błędu 400
url = "https://restcountries.com/v3.1/all?fields=name,capital,region,subregion,population,area,currencies"
response = requests.get(url)
data = response.json()

print(f"Pobrano dane dla {len(data)} krajów.")
```

```python
# Funkcja z podpowiedzi prowadzącego
def get_currency(currencies_dict):
    """Wyciąga kod pierwszej waluty z zagnieżdżonego słownika."""
    if currencies_dict:
        return list(currencies_dict.keys())[0]
    return None

kraje_lista = []

for kraj in data:
    # Pobieramy stolicę bezpiecznie - sprawdzamy czy lista nie jest pusta
    lista_stolic = kraj.get("capital", [])
    stolica = lista_stolic[0] if len(lista_stolic) > 0 else None
    
    kraje_lista.append({
        "nazwa": kraj.get("name", {}).get("common"),
        "stolica": stolica,
        "region": kraj.get("region"),
        "subregion": kraj.get("subregion"),
        "populacja": kraj.get("population"),
        "powierzchnia": kraj.get("area"),
        "waluta": get_currency(kraj.get("currencies"))
    })

# Tworzenie DataFrame
df = pd.DataFrame(kraje_lista)

# Wymagane wyświetlenie info o DataFrame
print("Kształt danych (shape):", df.shape)
print("\nTypy danych (dtypes):\n", df.dtypes)
print("\nPierwsze 5 wierszy (head):")
display(df.head())
```

```python
# Zapis do bazy danych
conn = sqlite3.connect("kraje_swiata.db")
df.to_sql("kraje", conn, if_exists="replace", index=False)
print("Dane zapisane do tabeli 'kraje' w bazie 'kraje_swiata.db'.")
```

```python
# 1. Jaka jest łączna populacja świata?
q1 = "SELECT SUM(populacja) AS laczna_populacja_swiata FROM kraje"
print(pd.read_sql_query(q1, conn))
print("-" * 30)

# 2. Które 10 krajów ma największą populację?
q2 = "SELECT nazwa, populacja FROM kraje ORDER BY populacja DESC LIMIT 10"
print(pd.read_sql_query(q2, conn))
print("-" * 30)

# 3. Ile krajów jest w każdym regionie i jaka jest ich średnia populacja?
q3 = "SELECT region, COUNT(*) AS liczba_krajow, AVG(populacja) AS srednia_populacja FROM kraje GROUP BY region"
print(pd.read_sql_query(q3, conn))
print("-" * 30)

# 4. Które kraje mają powierzchnię większą niż Polska (~312 679 km²)?
q4 = "SELECT nazwa, powierzchnia FROM kraje WHERE powierzchnia > 312679"
print(pd.read_sql_query(q4, conn))
print("-" * 30)

# 5. Który kraj ma najwyższą gęstość zaludnienia (populacja / powierzchnia)?
q5 = """
SELECT nazwa, (populacja / powierzchnia) AS gestosc_zaludnienia 
FROM kraje 
WHERE powierzchnia > 0 
ORDER BY gestosc_zaludnienia DESC 
LIMIT 1
"""
print(pd.read_sql_query(q5, conn))
```

```python
# Pobranie danych do wykresu
query_wiz = "SELECT region, SUM(populacja) AS populacja_regionu FROM kraje GROUP BY region"
df_wiz = pd.read_sql_query(query_wiz, conn)

# Prosty wykres słupkowy
df_wiz.plot(kind='bar', x='region', y='populacja_regionu', figsize=(10, 6), color='skyblue', legend=False)
plt.title("Łączna populacja każdego regionu")
plt.xlabel("Region")
plt.ylabel("Populacja")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Zamknięcie połączenia
conn.close()
```
